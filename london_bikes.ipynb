specifying the column names that i want to use

```python
cols_dict = {
    "timestamp": "time",
    "cnt": "count",
    "t1": 'real_temp_C',
    "t2": "feels_like_temp_C",
    "hum": "humidity_percent",
    "wind_speed": "wind_speed_kph",
    "weather_code": "weather",
    "is_holiday": "is_holiday",
    "is_weekend": "is_weekend",
    "season": "season"
}


# renaming the columns to the specified column name
london_bikes.rename(cols_dict, axis=1, inplace=True)
```

changing the humidity values to percentage (i.e. a value between 0 and 1)

```python
london_bikes
```

```python
london_bikes.humidity = london_bikes.humidity / 100
```

```python
# creating a season dictionary so that we can map the integers 0-3 to actual the written values
season_dict = {
    "0.0": "spring",
    "1.0": "summer",
    "2.0": "autumn",
    "3.0": "winter"
}



# creating a weather dictionary so that we can map the integers to actual the written values
weather_dict = {
    "1.0": "Clear",
    "2.0": "Scattered Clouds",
    "3.0": "Broken Clouds",
    "4.0": "Cloudy",
    "7.0": "Rain",
    "10.0": "Rain with thunderstorm",
    "26.0": "Snowfall"
}



# changing column data type of season to string
london_bikes.season = london_bikes.season.astype('str')
# mapping the values 0-3 to the actual written seasons
london_bikes.season = london_bikes.season.map(season_dict)



# changing column data type of season to string
london_bikes.weather = london_bikes.weather.astype('str')
# mapping the values 0-3 to the actual written seasons
london_bikes.weather = london_bikes.weather.map(weather_dict)
```
